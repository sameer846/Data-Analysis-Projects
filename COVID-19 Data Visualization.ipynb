```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
who_time_series = pd.read_csv("WHO.csv")
who_time_series.head()
```

```python
who_time_series.dtypes
```

```python
who_time_series["Date_reported"] = pd.to_datetime(who_time_series["Date_reported"])
```

```python
who_time_series.dtypes
```

```python
italy = who_time_series[who_time_series["Country"] == "Italy"]
```

```python
italy
```

```python
plt.plot(italy["Date_reported"], italy["Cumulative_cases"])

plt.show()
```

```python
plt.plot(italy["Date_reported"], italy["Cumulative_cases"])
plt.title("Italy : Cumulative Reported Cases")
plt.xlabel("Date")
plt.ylabel("Number of Cases")
plt.show()

print("Generally, a quantity that increase very quickly in the begining and the it slows down more and more over time has a LOGARITHMIC GROWTH")
```

```python
india = who_time_series[who_time_series["Country"] == "India"]
```

```python
plt.plot(india["Date_reported"], india["Cumulative_cases"])
plt.title("India : Cumulative Reported Cases")
plt.xlabel("Date")
plt.ylabel("Number of Cases")
plt.show()
```

```python
def plot_cases(country_name):
    df = who_time_series[who_time_series["Country"] == country_name]
    plt.plot(df["Date_reported"], df["Cumulative_cases"])
    plt.title("{} : CovidReported Cases".format(country_name))
    plt.xlabel("Date")
    plt.ylabel("Number of Cases")
    plt.ticklabel_format(axis = "y", style = "plain")
    plt.show()
```

```python
plot_cases("Poland")

print("Generally, a quantity that increase constantly over time has linear growth.")
```

```python
countries = who_time_series["Country"].unique()
countries[:10]
```

```python
for i in countries[:10]:
    plot_cases(i)
```

```python
france = who_time_series[who_time_series["Country"] == "France"]
uk = who_time_series[who_time_series["Country"] == "The United Kingdom"]
pakistan = who_time_series[who_time_series["Country"] == "Pakistan"]
```

```python
plt.plot(france["Date_reported"], france["Cumulative_cases"])
plt.plot(uk["Date_reported"], uk["Cumulative_cases"])
plt.show()

plt.plot(pakistan["Date_reported"], pakistan["Cumulative_cases"])
plt.show()
```

```python
plt.plot(france["Date_reported"], france["Cumulative_cases"], label = "France", color = "Blue")
plt.plot(uk["Date_reported"], uk["Cumulative_cases"], label = "UK", color = "red")
plt.plot(pakistan["Date_reported"], pakistan["Cumulative_cases"], label = "Pakistan", color = "green")
plt.legend()
plt.show()
```

# Scatter Plots and Correlations

```python
bike_sharing = pd.read_csv("day.csv")
```

```python
bike_sharing.head()
```

```python
bike_sharing["dteday"] = pd.to_datetime(bike_sharing["dteday"])
```

```python
bike_sharing["dteday"]
```

```python
# plt.figure(figsize=(10,4))
plt.plot(bike_sharing["dteday"], bike_sharing["cnt"])
plt.xticks(rotation = 90)

plt.show()
```

```python
# plt.figure(figsize=(10,4))
plt.plot(bike_sharing["dteday"], bike_sharing["temp"])
plt.xticks(rotation = 90)

plt.show()
```

```python
plt.plot(bike_sharing["dteday"], bike_sharing["windspeed"])
plt.xticks(rotation = 90)

plt.show()
```

```python
bike_sharing.temp.corr(bike_sharing.cnt)
```

```python
bike_sharing.temp.corr(bike_sharing.cnt)*100
```
